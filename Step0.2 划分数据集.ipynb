```python
import os
import pandas as pd
from glob import glob
root = r'F:\20230608-XueYuHang\Pathology\patches_norm'

patches = pd.DataFrame(glob(os.path.join(root, '*', '*.jpg')), columns=['fpath'])
patches['ID'] = patches['fpath'].map(lambda x: os.path.basename(os.path.dirname(x)).split('_')[0][:12])
patches['filename'] = patches['fpath'].map(lambda x: os.path.basename(x))
patches
```

```python
patch_info = pd.read_csv(os.path.join(root, '../patches/patch_info.csv'))
patch_info = patch_info[patch_info['label_id'] == 1]
patch_info['filename'] = patch_info['filename'].map(lambda x: os.path.basename(x))
patch_info
```

```python
patches = pd.merge(patches, patch_info['filename'], on='filename', how='inner')
patches
```

```python
patches.to_csv('data/path_info.csv', index=False)
```

# 正式划分数据

```python
import pandas as pd
from onekey_algo.custom.components.metrics import get_time_dependent_gt

label_data = pd.read_csv('data/survival.csv')
label_data = get_time_dependent_gt(label_data, time=365, duration_col='OSTime', event_col='OS')
print(label_data['label'].value_counts())
label_data
```

```python
import pandas as pd
import os
import numpy as np
from onekey_algo.custom.components.comp2 import split_dataset4sol
from onekey_algo import get_param_in_cwd

root = get_param_in_cwd('data_root')
ds = split_dataset4sol(label_data, y_data=label_data['label'], n_trails=5, save_dir=os.path.join(root), 
                       random_state=0, cv=False, test_size=0.33)

for idx, (train, val) in enumerate(ds):
    train = pd.merge(train, patches[['ID', 'fpath']], on='ID', how='inner')
    train[['fpath', 'label']].to_csv(os.path.join(root, 'split_info', f'train-RND-{idx}.txt'), index=False, header=False, sep='\t')
    val = pd.merge(val, patches[['ID', 'fpath']], on='ID', how='inner')
    val[['fpath', 'label']].to_csv(os.path.join(root, 'split_info', f'val-RND-{idx}.txt'), index=False, header=False, sep='\t')
    print(f"随机划分：{idx}，训练集：{train.shape[0]}, {len(np.unique(train['ID']))}, 测试集集：{val.shape[0]}, {len(np.unique(val['ID']))}")
```
